Enfoque

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("gato.jpg")

mascara = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

img_fin = cv2.filter2D(img, -1, mascara)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_fin_rgb = cv2.cvtColor(img_fin, cv2.COLOR_BGR2RGB)

plt.subplot(121)
plt.imshow(img_rgb)
plt.title("Original")

plt.subplot(122)
plt.imshow(img_fin_rgb)
plt.title("Filtrada")

plt.show()
```
